```python
import time
from datetime import datetime, timedelta
```

```python
def timer_dec(base_fn):
    def enhanced_fn(*args, **kwargs):
        start_time = time.time()
        result = base_fn(*args,**kwargs)
        end_time = time.time()
        print(f"Task time: {end_time - start_time} seconds \n")
        return result
    return enhanced_fn

@timer_dec
def brew_tea(tea_type, sleep_time):
    print(f"Brewing {tea_type} tea...")
    time.sleep(sleep_time)
    print("Tea is ready!")

@timer_dec
def make_matcha():
    print("Making Matcha...")
    time.sleep(1)
    print("Matcha is ready!")
    return f"Drink matcha by {datetime.now() + timedelta(minutes=30)}"

brew_tea(tea_type="green", sleep_time=1)
print(make_matcha())
```

Problem 1
Create a decorator called mission_timer that:

Prints a 3 → 2 → 1 → Launch! countdown
Times how long the mission function takes
Works with two functions that take no arguments and return nothing
Then decorate and call:

launch_probe()
deploy_satellite()

```python
def mission_timer(base_fun):
    def enhanced_fn(*args,**kwargs):
        #countdown
        for i in range(3, 0 , -1):
            print(i)
            time.sleep(1)
        print("Launch \n")
        #mission time
        start_time = time.time()
        result = base_fun(*args,**kwargs)
        end_time = time.time()
        print(f"Task time : {end_time - start_time} seconds \n")
        return result
    return enhanced_fn


@mission_timer
def launch_probe(target):
    print(f"Launching probe toward {target}...!")
    time.sleep(1)
    return f"Probe successfully en route to {target}."

@mission_timer
def deploy_satellite(orbit_type,altitude_km):
    print(f"Deploying satellite into {orbit_type} orbit at {altitude_km} km...")
    time.sleep(2)

```

```python
launch_probe("Mars")
deploy_satellite("polar", 800)
deploy_satellite(orbit_type="geostationary", altitude_km=35786)
```
