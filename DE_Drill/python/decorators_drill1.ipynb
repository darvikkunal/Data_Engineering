Python — Decorator Practice

Step 1 — understand this first before coding:

A decorator is just a function that wraps another function. The skeleton ALWAYS looks like this:

```python
def my_decorator(func):        # takes a function
    def wrapper(*args, **kwargs):
        # do something BEFORE
        result = func(*args, **kwargs)  # run original function
        # do something AFTER
        return result
    return wrapper              # return wrapper WITHOUT ()

@my_decorator
def my_function():
    pass
```

Now your exercise:

Write a decorator called validate_age that:

Checks if the age argument passed to a function is valid (between 0 and 120)

If valid → runs the function normally

If invalid → prints "Invalid age!" and does NOT run the function

Apply it to a function called register_user(name, age) that prints "User {name} registered with age {age}".

```python
def validate_age(func):
    def wrapper(*args,**kwargs):
        age = args[1]
        if 0 <= age <= 120:
            result = func(*args,**kwargs)
            print("Valid Age")
            return result
        else:
            print("Invalid Age")
            return
    return wrapper

@validate_age
def register_user(name,age):
    print(f"User : {name} registered with age {age}")
```

```python
register_user("Kunal", 25)    # should work
register_user("Rahul", -5)    # should print Invalid age!
register_user("Anjali", 150)  # should print Invalid age!
```
