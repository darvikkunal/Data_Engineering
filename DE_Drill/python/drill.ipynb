Create a program that takes a user's full name as input and prints:

The name in ALL CAPS
The name in all lowercase
The total number of characters (excluding spaces)
The name in reverse order

```python
name = input("Enter the name : ")
print(f"Name in caps : {name.upper()}")
print(f"Name in lower : {name.lower()}")
print(f"Length of name : {(len(name.strip()))}")
print(f"Name in reverse order : {name[::-1]}")
```
