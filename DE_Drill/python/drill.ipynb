Create a program that takes a user's full name as input and prints:

The name in ALL CAPS
The name in all lowercase
The total number of characters (excluding spaces)
The name in reverse order

```python
name = input("Enter the name : ")
print(f"Name in caps : {name.upper()}")
print(f"Name in lower : {name.lower()}")
print(f"Length of name (no spaces): {(len(name.replace(' ','')))}")
print(f"Name in reverse order : {name[::-1]}")
```

Python — Exercise 5

Create a program that asks the user to enter a string and prints:

Total number of vowels in the string
Total number of consonants in the string
Whether the string is a palindrome or not (reads same forwards and backwards, ignore case and spaces)

```python
vowels = 'aeiouAEIOU'
string = input("Enter the string : ")
vowels_count = 0
consonants_count = 0

# vowels
for char in string:
    if char in vowels:
        vowels_count += 1
    elif char.isalpha():
        consonants_count += 1

print(f" The Input String is : {string}")
print(f" The count of vowels : {vowels_count}")
print(f" The count of consonants : {consonants_count}")

# Palindrome
#cleaned string
clean_str = string.lower().replace(' ','')
#reverse str
reverse_str = clean_str[::-1]
if clean_str == reverse_str:
    print("Palindrome : YES")
else :
    print(f"Palindrome : NO")
```

Create a number guessing game. The program picks a random number between 1 and 100. The user keeps guessing until they get it right. After each wrong guess tell them if the answer is higher or lower. At the end tell them how many attempts it took.

Hint: use the random module — random.randint(1, 100)

```python
# Step 1 — import random and generate secret number
import random
secret = random.randint(1, 100)

# Step 2 — set up attempt counter
attempts = 0

# Step 3 — while loop that keeps running until guess is correct
while True:
    try:
        guess = int(input("Enter your guess: "))
    except ValueError:
        print("Please enter a valid number")
        continue
    attempts += 1  # increment counter every guess
    
    # Step 4 — inside the loop:
    # if guess is too low  → print "Go Higher"
    if guess < secret:
        print("Guess a higher number")
    # if guess is too high → print "Go Lower"
    elif guess > secret:
        print("Guess a lower number")
    # if guess is correct  → print success + attempts
    else:
        print(f"You have guessed correctly in {attempts} attempts")
        break

    #                      → BREAK out of the loop break
```

Write a program to print the Fibonacci Sequence

```python
def fibonacci(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
    
fibonacci_number = int(input("Enter the number : "))
print(f"Printing the first {fibonacci_number} terms : ")

for i in range(fibonacci_number):
    print(fibonacci(i), end=' ')
    
```
