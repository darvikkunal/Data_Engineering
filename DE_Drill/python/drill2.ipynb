Python — Exercise 3: datetime module

Create a program that:

Prints today's date and time

Prints the day of the week (e.g. Monday)

Asks the user to enter a date in DD-MM-YYYY format and tells them how many days ago or how many days from now that date is

Prints what the date will be 90 days from today

```python
from datetime import datetime , timedelta
today = datetime.now()

print(f"Today's datetime : {today}")


print(f"Today's date : {today.strftime('%d-%m-%Y')}")


print(f"Today's day : {today.strftime('%A')}")

```

```python
user_input = input('Enter date in [DD-MM-YYYY] format: ')
user_date = datetime.strptime(user_input,'%d-%m-%Y')
date_diff = (user_date - today).days
print(f"User Entered date : {user_date}")

if today < user_date:
    print(f"{date_diff} days from today")
else:
    print(f"{abs(date_diff)} days ago from today")


print(f"90 days from today will be : {today + timedelta(days=90)}")
```

🐍 Python — Exercise 14: json module
Create a program that:

Has a list of dictionaries representing orders (at least 3 orders, each with order_id, customer, amount, status)

Writes this data to a file called orders.json

Reads it back from the file

Prints only orders where status == "pending"

Prints the total amount of all pending orders

```python
import json
file_path = '/Users/darvikkunalbanda/DataEngineering/DE_Drill/dataset/orders.json'
```

```python
orders = [
        {"order_id": 101,"customer": "Rahul Sharma","amount": 2500,"status": "Delivered"},
        {"order_id": 102,"customer": "Anjali Verma","amount": 1800,"status": "Pending"},
        {"order_id": 103,"customer": "Kiran Reddy","amount": 3200,"status": "Shipped"},
        {"order_id": 104,"customer": "Nilo Singh","amount": 2800,"status": "Pending"}
]
```

```python
with open(file_path,'w') as f:
    json.dump(orders,f,indent=4)
```

```python
with open(file_path,'r') as f:
    data = json.load(f)

    filtered_data = [x for x in data if x['status'].lower() == 'pending']
    print(filtered_data)
    total_amount = sum([x['amount'] for x in filtered_data])
    print(f"The total amount of Pending orders : {total_amount}")
```
