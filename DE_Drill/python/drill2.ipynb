Python — Exercise 3: datetime module

Create a program that:

Prints today's date and time

Prints the day of the week (e.g. Monday)

Asks the user to enter a date in DD-MM-YYYY format and tells them how many days ago or how many days from now that date is

Prints what the date will be 90 days from today

```python
from datetime import datetime , timedelta
today = datetime.now()

print(f"Today's datetime : {today}")


print(f"Today's date : {today.strftime('%d-%m-%Y')}")


print(f"Today's day : {today.strftime('%A')}")

```

```python
user_input = input('Enter date in [DD-MM-YYYY] format: ')
user_date = datetime.strptime(user_input,'%d-%m-%Y')
date_diff = (user_date - today).days
print(f"User Entered date : {user_date}")

if today < user_date:
    print(f"{date_diff} days from today")
else:
    print(f"{abs(date_diff)} days ago from today")


print(f"90 days from today will be : {today + timedelta(days=90)}")
```

Python — Exercise 16: OOP / Classes
Create a class called BankAccount that:

Has attributes: owner, balance
Has methods:

deposit(amount) — adds to balance, prints new balance

withdraw(amount) — deducts from balance, but prints a warning if insufficient funds

get_balance() — prints current balance


Create 2 different account objects and test all methods
        

```python
class BankAccount:
    def __init__(self, owner , balance = 0):
        self.owner = owner
        self.balance = balance
    
    def deposit(self, amount):
        self.balance += amount
        print(f"The new deposit balance is : {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print(f"Insufficient funds! Balance is only {self.balance}")
        else:
            self.balance -= amount
            print(f"The withdrawal balance is : {self.balance} \n withdrawn amount is {amount}")

    def get_balance(self):
        print(f"{self.owner}'s balance: {self.balance}")


account1 = BankAccount('kunal',5000)
acccount2 = BankAccount('Jules',6000)

account1.deposit(3000)
account1.withdraw(2000)

print("\n")

acccount2.withdraw(1500)
acccount2.deposit(2000)
```

**Exercise A — Student**

- Attributes: name, grade, marks

- Methods: add_marks(subject, score), get_average(), is_passing() — returns True if average > 40

```python
class student:
    def __init__(self,name,grade,marks=0):
        self.name = name
        self.grade = grade
        self.marks = []

    def add_marks(self,subject,marks):
        self.marks.append(marks)
        print(f"You have got {marks} marks in {subject}")

    def get_avg(self):
        avg = float(sum(self.marks)/ len(self.marks))
        #print(f"The average is {avg}")
        return avg
        
    
    def is_passing(self):
        avg = self.get_avg()
        if avg > 40:
            print(f"{self.name}'s average is : {avg} So, PASSED")
        else:
            print(f"{self.name}'s average is : {avg} So, FAILED")

    

st1 = student('kili','A',)
st2 = student('Miro','B',)

st1.add_marks('Social',49.5)
st1.add_marks('Math',49.1)
st1.get_avg()
st1.is_passing()
print('\n')
st2.add_marks('Social',29.5)
st2.add_marks('Math',69.1)
st2.get_avg()
st2.is_passing()
```

**Exercise B — Product**

- Attributes: name, price, stock

- Methods: restock(quantity), sell(quantity) — warn if out of stock, get_info()

```python
class product():
    def __init__(self,name,price,stock=0):
        self.name = name
        self.price = price
        self.stock = stock

    def restock(self,quantity=0):
        self.stock += quantity
        print(f"The quantity of the restock is {self.stock}")

    def sell(self,quantity=0):
        if quantity > self.stock:
            print(f"Out of stock! only {self.stock} left")
        else:
            self.stock -= quantity
            print(f"Sold {quantity} | Remaining Stock : {self.stock}")

    def get_info(self):
        print(f"The product name : {self.name} | The product price : {self.price} | The product stock : {self.stock} ")
        

p1 = product('bike_toy',45.90,10)
p2 = product('car_toy',69.61,8)
p3 = product('sugar',56.70,5)

p1.restock(2)
p1.sell(1)
p1.get_info()
print('\n')

p2.restock(4)
p2.sell(2)
p2.get_info()
print('\n')

p3.restock(-2)
p3.sell(100)
p3.get_info()
print('\n')

```

**Exercise C — Employee**

- Attributes: name, role, salary

- Methods: give_raise(percent), get_details(), promote(new_role)

```python
class employee():

    def __init__(self,name,role,salary=1000):
        self.name = name
        self.role = role
        if salary < 0:
            raise ValueError("Salary Cannot be negative")
        self.salary = salary

    def give_raise(self,percent=2):
        self.salary = self.salary * (1 + percent/100)
        print(f"{self.name} got a raise of {percent}% | NEW SALARY IS : ${self.salary}")

    def get_details(self):
        print(f"Employee Name : {self.name.upper()} | Employee Role : {self.role} | Employee Salary : ${self.salary}")

    def promoted(self,new_role):
        self.role = new_role
        print(f"CONGRATS! You've got Promoted as {self.role}")

```

```python
emp1 = employee("KILO", "Data Engineer",3000)
emp1.get_details()
emp1.give_raise()
emp1.promoted("Senior Data Engineer")
emp1.give_raise(-10)
emp1.get_details()
print('\n')

emp2 = employee("Rahul", "Analyst")
emp2.get_details()
emp2.give_raise(10)
print('\n')

emp3 = employee("Anjali", "Manager", 10000)
emp3.give_raise(10)   # 11000
emp3.give_raise(8)   # 12100
emp3.get_details()
print('\n')

emp4 = employee("Zero", "Intern", 500)
emp4.give_raise(10)
emp4.give_raise(200)
emp4.promoted("SE")
emp4.get_details()
```

Python — Exercise 18: CSV module
Create a program that:

Creates a list of employee dicts (name, role, salary)

Writes it to employees.csv using csv.DictWriter

Reads it back using csv.DictReader

Prints only employees with salary > 50000

Prints the average salary of all employees

```python
employees = [
    {"name": "Kiju", "role": "Data Engineer", "salary": 80000},
    {"name": "Rahul", "role": "Data Analyst", "salary": 6000},
    {"name": "Anjali", "role": "ML Engineer", "salary": 9000},
    {"name": "Sneha", "role": "BI Analyst", "salary": 55000},
    {"name": "Arjun", "role": "Backend Engineer", "salary": 75000}
]
```

```python
import csv

#writing
with open('employees.csv', 'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['name','role','salary'])
    writer.writeheader()
    writer.writerows(employees)
```

```python
#reading
salaries = []
with open('employees.csv','r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        salary = int(row['salary'])
        salaries.append(salary)
        if salary > 50000:
            print(f" The salaries above 50000 : {[row['name'], row['salary']]}")


avg_salary = sum(salaries) / len(salaries)
print(f"Average Salary : {avg_salary}")
```
