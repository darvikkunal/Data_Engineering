Python — Pandas Exercise 2

Using Pandas, read orders.csv and order_details.csv. Then:

Merge them on OrderID
Create a new column called line_revenue = UnitPrice * Quantity * (1 - Discount)
Find the total revenue per order (group by OrderID, sum line_revenue)
Print the top 5 orders by revenue

```python
ordercsv = pd.read_csv('/Users/darvikkunalbanda/DataEngineering/DE_Drill/dataset/northwind_data/orders.csv')
order_detailscsv = pd.read_csv('/Users/darvikkunalbanda/DataEngineering/DE_Drill/dataset/northwind_data/order_details.csv')
```

```python
print(ordercsv.columns)
print(order_detailscsv.columns)
```

```python
merged_df = pd.merge(ordercsv,order_detailscsv, on='orderID')
```

```python
merged_df['line_revenue'] = merged_df['unitPrice'] * merged_df['quantity'] * (1 - merged_df['quantity'])
```

```python
merged_df.groupby('orderID')['line_revenue'].sum().sort_values(ascending=False).head()
```
