Validate the quality and correctness of email values:
- Must not be empty
- Must contain '.' and '@'
- Must contain exactly one '@' symbol
- Must end with '.com', '.org', or '.net'
- Must not be longer than 254 characters
- Must start and end with a letter or digit

```python
email = input("Enter you email : ")
valid = True
if email.strip() == "":
    print("Email is empty")
    valid=False
if not('.' in email and '@' in email):
    print("Email must contain '.' and '@' ")
    valid=False
if email.count('@') != 1:
    print("Email must contain exactly one '@' ")
    valid=False
if not email.endswith(('.com' or '.org' or '.net')):
    print("Email Must end with '.com', '.org', or '.net' ")
    valid=False
if len(email) > 254:
    print("Email Must not be longer than 254 characters")
    valid=False
if not (email[0].isalnum() and email[-1].isalnum()):
    print("Email Must start and end with a letter or digit ")
    valid=False
if valid:
    print("Email is valid")
```

Validate the quality and correctness of passwords:
- Must not be empty
- Must be at least 8 characters
- Must include at least 1 uppercase
- Must include at least 1 lowercase
- Must not be same as the email
- Must not contain any spaces
- Must start and end with a letter or digit

```python
password = input("Enter your password : ")
email = "kl.kl.com"
valid = True
if password.strip()=="":
    print("password is empty")
    valid = False
if len(password) < 8:
    print("Password Must be at least 8 characters")
    valid = False
if password == password.lower():
    print("Password Must include at least 1 uppercase")
    valid = False
if password == password.upper():
    print("Password Must include at least 1 lowercase")
    valid = False
if password==email:
    print("Password Must not be same as the email")
    valid = False
if " " in password:
    print("Password must not contain spaces")
    valid = False
if not (password[0].isalnum() and password[-1].isalnum()):
    print("Password Must start and end with a letter or digit ")
    valid=False
if valid:
    print("password is valid")
```

Username Rules

Your program should check all of the following:

Username must not be empty

Username must be between 6 and 15 characters

Username must contain at least 1 lowercase letter

Username must contain at least 1 digit

Username must not start with a digit

Username must not contain spaces

Username must not be the same as the word "admin"

Username must end with a letter

```python
admin = input("Enter admin: ")
username = input("Enter Username : ")

valid = True
if username.strip()=="":
    print("Username must not be empty")
    valid=False
if len(username) <6 and len(username) > 15:
    print("Username must be between 6 and 15 characters")
    valid=False
if username.isalpha() ==1:
    print("Username must contain at least 1 digit")
    valid=False
if username[0].isdigit():
    print("Username must not start with a digit")
    valid=False
if " " in username:
    print("Username must not contain spaces")
    valid=False
if username==admin:
    print(f"Username must not be the same as the word {admin} ")
    valid=False 
if not username[-1].isalpha():
    print("Username must end with a letter")
    valid = False
if valid:
    print("valid")
```

```python
# Extend the above code to 
#1. Allow up to 3 attempts
#2. If the user types "yes", print "Glad we are on the same page"
#3. Otherwise, print "3 Strikes, you are out!"
```

```python
attempts = 0
max_attempts = 3

while attempts < max_attempts:
    ans = input("Do you agree? (yes/no): ").lower()

    if ans == "yes":
        print("Glad we are on the same page")
        break
    attempts += 1
else:
    print("3 strikes, you are out!")
```
