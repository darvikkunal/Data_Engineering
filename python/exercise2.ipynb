### Copy a list

```python
# reference
# Create a copy of the list in a new variable
letters = ['a','b','c']
letters_copy = letters
letters.pop()
letters_copy.append('z')
print(f"original list : {letters_copy}")
print(f"copy : {letters_copy}")
```

```python
#shallow copy
# Create a copy of the list in a new variable
letters1 = ['a','b','c']
letters_copy1 = letters1.copy()
letters1.pop()
letters_copy1.append('z')
print(f"original list : {letters1}")
print(f"copy : {letters_copy1}")
```

```python
matrix = [
    ['d', 'e'],
    ['a', 'b',],   # Row 1
]
matrix_copy = matrix.copy()
matrix.pop()
matrix_copy[0].append('z')
print('original : ',matrix)
print('copy : ',matrix_copy)
```

```python
import copy
matrix1 = [
    ['d', 'e'],
    ['a', 'b',],   # Row 1
]

matrix_copy12 = copy.deepcopy(matrix1)
matrix1.pop()
matrix_copy12[0].append('z')
print('Original : ',matrix1)
print('copy : ',matrix_copy12)

```

```python
#IS OPERATOR

original = [
    ['a','b'],
    ['c','d'],
]

#Assignment
copy1 = original
print("Same object?", original is copy1, "\n")

#Shallow copy
copy2 = original.copy()
print("same object?", original is copy2)
print("Shared lists?", original[0] is copy2[0], "\n")

# Deep Copy
copy3 = copy.deepcopy(original)
print("same object?", original is copy3)
print("Shared Lists?", original[0] is copy3[0], "\n")
```

```python
# lambda
multiple = lambda x: x*2
print(multiple(5))

addl = lambda x,y:x+y
print(addl(5,6))

check = lambda i: i in 'python'
print(check('z'))
```

```python
#lambda + map

price = ['$12.50','$9.99','$100.00']
#p='$9.99'
#float(p.replace('$',''))

print(list(map(lambda p:float(p.replace('$','')),price)))
```

```python
#lambda + filter

prices = [120,40,500,23,540,546,23]

print(list(filter(lambda p:p >= 100,prices)))
```

```python
#lambda + filter

students = [
    ['Maria', 85],
    ['Kumar', 90],
    ['Max', 60]
]

print(list(filter(lambda row:row[1] > 70,students)))
```

```python
print(list(filter(lambda x:x[0].lower().startswith('k'),students)))
```

```python
#lambda + sort
print(list(sorted(students, key=lambda x:x[1], reverse=True)))
```
