### Lists

```python
#create Lists
empty = []
letters = ['a','b','c']
numbers = [1,2,4]
mixed = [1, 'a',True,None]
print(empty)
print(type(empty))
print(letters)
print(numbers)
print(mixed)
```

```python
empty = list()
print(empty)

letters = list('python')
print(letters)

numbers = list(range(5))
print(numbers)
```

```python
# Nested List
matrix = [['a','b','c'],
          ['d','e','f']]

mixed_matrix = [['a','b'],
                [1,2,4],
                [True]]
print(matrix)
print(mixed_matrix)
```
