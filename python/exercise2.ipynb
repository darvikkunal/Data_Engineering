### Lists

```python
#create Lists
empty = []
letters = ['a','b','c']
numbers = [1,2,4]
mixed = [1, 'a',True,None]
print(empty)
print(type(empty))
print(letters)
print(numbers)
print(mixed)
```

```python
empty = list()
print(empty)

letters = list('python')
print(letters)

numbers = list(range(5))
print(numbers)
```

```python
# Nested List
matrix = [['a','b','c'],
          ['d','e','f']]

mixed_matrix = [['a','b'],
                [1,2,4],
                [True]]
print(matrix)
print(mixed_matrix)
```

```python
lst = ['a','b','c','d']
lst[0]
lst[-1]
lst[3]
```

```python
# Access & Read
matrix = [
    ['a', 'b', 'c'],   # Row 0
    ['d', 'e', 'f'],   # Row 1
    ['g', 'h', 'i']    # Row 2
]
print(matrix[0][1])
print(matrix[-1][0])
print(matrix[:][:-1])
print(matrix[2][:2])
```

#### Unpacking

```python
person = ['maria', 29 , 'Data Engineer', 'Spain']

name , age , role , country = person

print(role)
```

```python
person = ['maria', 29 , 'Data Engineer', 'Spain']

name , *details, country = person

print(details)
```

```python
person = ['maria', 29 , 'Data Engineer', 'Spain']

name , age , _ , country = person

print(_)
```

```python
person1 = ['maria', 29 , 'Data Engineer', 'Spain']

a ,  *_ , b = person1

print(a)
print(b)
```
