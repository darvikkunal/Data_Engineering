4. Match‑case menu
Implement:

python
print("1) Load data")
print("2) Run quality checks")
print("3) Exit")
choice = input("Choose option: ")

match choice:
    case "1":
        print("Loading data...")
    case "2":
        print("Running quality checks...")
    case "3":
        print("Goodbye")
    case _:
        print("Invalid choice")
Then wrap this in a while loop so the menu repeats until the user picks 3.

```python
while True:
    print("1) Load data")
    print("2) Run quality checks")
    print("3) Exit")
    choice = input("Choose option: ")
    match choice:
        case "1":
            print("Loading data...")
        case "2":
            print("Running quality checks...")
        case "3":
            print("Goodbye")
            break
        case _:
            print("Invalid choice")
```

5. For / while + break / continue / for‑else
Given a list of transaction amounts:

python
amounts = [100, 50, -20, 30, 0]
Loop and stop at the first negative using break.

If you finish the loop with no negative, print "All non‑negative" using for‑else.

```python
amounts = [100, 50, -20, 30, 0]
found_negative = False
total = 0

while True:
    for amount in amounts:
        if amount <= 0:
            continue

        if amount <0:   
            print(f"Negative found : {amount}")
            found_negative = True
            break
        total += amount
    else:
        print(f"All are +ve numbers : {amounts}")
        print("Sum (zeros skipped): ",total)
    break
```
